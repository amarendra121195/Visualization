# 1. scatter plots
 The scatter plot is a mainstay of statistical visualization. It depicts the joint  distributions of two variables using a cloud of points, where each point represents an observations int the datasets.
 This depication allows the eye to infer substantial about where there is any meningful relatioonship between them.
 
 > scatter plot can be plotted for both numerical as well as categorical data, its mostly used
 to depict relationship between numerical data.
 
 There are 2 ways to draw a scatterplot in seaborn
 
 1. relpot[Figur Level Function]
 2. scatterplot[Axes Level Function]
 
 
 As we will see, these visualizations can be quite helpful because they use simple and easily-
 understood representations of data even on complex data by mapping up to three additinal
 variables using the semantics of hue , size, and style.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.show()
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

# plot a Scatter plot between total_bill and tip

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

# The hue parameters

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

# The style parameters

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

# The size Parameters

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

# Using the scatterplot function

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
```

# Jointplot
Jointplot creats a multi-panel figure that shows both the bivariate (or joint) relationship
two variables along with the univariate (or marginal) distributions of each on separate axes.

We can use jointplot in the following 4 manners:

    1. scatterplots
    2. Hexbin plots
    3. kdeplots
    4. regplot

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips) #hex is the 2d histogram 
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='reg',data=tips)
```

# Distplots

Distplots are used to do univariate on a continous / numerical data

Distplot combines 3 things in one graph

1. Histogram
2. Kernel Density Enstimations(PDF)
3. Rugplot

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().mean()
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df.isnull().mean()
```

# Plotting a Displot

```python
sns.distplot(df['age'])
```

# Plotting only the Histogram
> counting the no of observation in a given bins

```python
sns.distplot(df['age'],bins=10,kde=False)
```

```python
# Plotting only the kde
sns.distplot(df['age'],hist=False)
```

```python
# Plotting only the rugplot
# every vertical line 
sns.distplot(df['age'],hist=False,kde=False,rug=True)
```

# How to use the Distplot

```python
sns.distplot(df[df['survived']==1]['age'])
sns.distplot(df[df['survived']==0]['age'])
```
