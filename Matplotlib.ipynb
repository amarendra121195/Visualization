```python
import matplotlib.pyplot as plt
%matplotlib inline     # print image
```

# pie chart

```python
areas=['marketing','sales','development','hr','customer support']
budget=[2.5,4.5,10,1,6]
```

```python
plt.pie(budget,labels=areas,autopct='%0.2f%%')
plt.show()
```

```python
x=[23,45,12,49,97,32,11,44]
y=[45,34,67,87,55,22,89,90]
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('salary')
plt.title('age vs salary')
plt.show()
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,565,565,899]
chor_bazzar=[100,200,300,400,250,100,450,150]
```

```python
plt.plot(product_id,online_price,color='r',marker='+',markersize=10) # 2D plot (list,numpy array)
plt.plot(product_id,offline_price,color='y',marker='+',markersize=10)
plt.plot(product_id,chor_bazzar,color='g',marker='+',markersize=10)
plt.xlabel('product_id')
plt.ylabel('online_price')
plt.title('product price comparision')
plt.show()  # information will be remove and get only graph
```
